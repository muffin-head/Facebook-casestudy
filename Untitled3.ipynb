```python
r=csv.reader(open('Downloads/Case Study 03/FacebookRecruiting/trainWOH.csv','r'))
edges=dict()
for i in r:
    edges[i[0],i[1]]=1
```

```python
import random
missing_edges=set([])
while (len(missing_edges) <9437519):
    a= random.randint(1,len(g.nodes()))
    b=  random.randint(1,len(g.nodes()))
    tmp=edges.get((a,b),-1)
    if tmp==-1 and a!=b:
        try:
            if nx.shortest_path_length(g,source=a,target=b)>2:
                missing_edges.add((a,b))
            else:
                continue
        except:
            missing_edges.add((a,b))
            
        
```

```python
from sklearn.model_selection import train_test_split

no_edges_list = pickle.load(open('Downloads/Case Study 03/FacebookRecruiting/missing_edges_final.p','rb'))
pos=pd.read_csv('Downloads/Case Study 03/FacebookRecruiting/train.csv')
neg=pd.DataFrame(list(no_edges_list),columns=['source_node', 'destination_node'])
X_train_pos, X_test_pos, y_train_pos, y_test_pos  = train_test_split(pos,np.ones(len(pos)),test_size=0.2, random_state=9)
X_train_neg, X_test_neg, y_train_neg, y_test_neg  = train_test_split(neg,np.zeros(len(neg)),test_size=0.2, random_state=9)
X_train_pos.to_csv('Downloads/Case Study 03/FacebookRecruiting/train_pos_after_eda.csv',header=False, index=False)
X_test_pos.to_csv('Downloads/Case Study 03/FacebookRecruiting/test_pos_after_eda.csv',header=False, index=False)
X_train_neg.to_csv('Downloads/Case Study 03/FacebookRecruiting/train_neg_after_eda.csv',header=False, index=False)
X_test_neg.to_csv('Downloads/Case Study 03/FacebookRecruiting/test_neg_after_eda.csv',header=False, index=False)
X_train_pos = pd.read_csv('Downloads/Case Study 03/FacebookRecruiting/train_pos_after_eda.csv', names=['source_node', 'destination_node'])
X_test_pos = pd.read_csv('Downloads/Case Study 03/FacebookRecruiting/test_pos_after_eda.csv', names=['source_node', 'destination_node'])
X_train_neg = pd.read_csv('Downloads/Case Study 03/FacebookRecruiting/train_neg_after_eda.csv', names=['source_node', 'destination_node'])
X_test_neg = pd.read_csv('Downloads/Case Study 03/FacebookRecruiting/test_neg_after_eda.csv', names=['source_node', 'destination_node'])
```

```python
X_train  = pd.concat([X_train_pos, X_train_neg], ignore_index=True)
y_train = np.concatenate((y_train_pos,y_train_neg))
X_test = pd.concat([X_test_pos, X_test_neg], ignore_index=True)
y_test = np.concatenate((y_test_pos,y_test_neg)) 
X_train.to_csv('Downloads/Case Study 03/FacebookRecruiting/train_after_eda.csv',header=False,index=False)
X_test.to_csv('Downloads/Case Study 03/FacebookRecruiting/test_after_eda.csv',header=False,index=False)
pd.DataFrame(y_train.astype(int)).to_csv('Downloads/Case Study 03/FacebookRecruiting/train_y.csv',header=False,index=False)
pd.DataFrame(y_test.astype(int)).to_csv('Downloads/Case Study 03/FacebookRecruiting/test_y.csv',header=False,index=False)
```
